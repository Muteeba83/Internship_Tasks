```python
#Load dataset
data=sns.load_dataset('iris')
```

```python
#columns name 
print('columns name in dataset :\n',data.columns)
```

```python
data.describe()
```

```python
#scatter plot to show relationships between features
graph=sns.scatterplot(data,x='sepal_length',y='sepal_width',hue='species',style='species',markers=['o', 's', 'D'],palette='bright',s=20)
plt.title('Scatter plot of sepal_length & sepal_width')
plt.show()
```

```python
#histograms to show value distributions
graph2=sns.histplot(data, x='sepal_length', bins=20, kde=False, hue='species')
plt.title('Histogram of sepal_length of different species')
plt.show()
```

```python
#box plots to identify outliers
graph3 = sns.boxplot(data, x='species', y='sepal_length', hue='species', dodge=True)
plt.title('Box Plot of sepal_length & sepal_width')
plt.show()
```
